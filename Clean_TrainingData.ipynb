```python
import numpy as np
import pandas as pd
```

```python
adit_df = pd.read_csv("dataset_adit.csv")
adit_df.sample(20)
```

```python
alus_df = pd.read_csv("dataset_alus.csv")
alus_df.sample(20)
```

```python
amin_df = pd.read_csv("dataset_amin.csv")
amin_df.sample(20)
```

```python
def assessing_data(df):
    print(f"\n---TABLE INFO---\n")
    df.info()

    print(f"\n---MISSING VALUES---\n")
    print(df.isna().sum())

    print(f"\n---DUPLICATE DATA---\n")
    print(df.duplicated().sum())

    print(f"\n---STATISTICS PARAMETER---\n")
    print(df.describe())
```

```python
assessing_data(adit_df)
```

```python
assessing_data(alus_df)
```

```python
assessing_data(amin_df)
```

```python
def preprocess_data(data):
  # Menambahkan kolom baru
  data['amount'] = data['MUTASI_DEBET'] + data['MUTASI_KREDIT']

  # Menghapus kolom tidak terpakai
  data.drop(["ID","NOREK","TGL_EFEKTIF","JAM_TRAN","SEQ","DESK_TRAN","SALDO_AWAL_MUTASI",
             "MUTASI_DEBET", "MUTASI_KREDIT", "SALDO_AKHIR_MUTASI","TRUSER","KODE_TRAN",
             "KODE_TRAN_TELLER","TRREMK", "TLBDS1","TLBDS2","REMARK_CUSTOM"],
            axis = 1, inplace= True)

  # Mengganti nama kolom
  data.rename(columns={'TGL_TRAN':'date','GLSIGN':'transaction_type'}, inplace=True)

  # Mengganti isi kolom
  data['transaction_type'] = data['transaction_type'].replace({'Cr': 'Income', 'Db': 'Expense'})

  # Mengkonversi tipe data
  data['date'] = pd.to_datetime(data['date']).dt.date
  data['amount'] = pd.to_numeric(data['amount'])
  data['transaction_type'] = data['transaction_type'].astype('category')

  # Menghapus baris dengan missing value
  data.dropna(inplace=True)

  # Mengurutkan data
  data.sort_values(by='date', ascending=True)

  return data
```

```python
preprocess_data(adit_df)
```

```python
preprocess_data(alus_df)
```

```python
preprocess_data(amin_df)
```

```python
data = pd.concat([adit_df, alus_df, amin_df]).sort_values(by='date', ascending=True)
data.to_csv("merger_dataset.csv", index=False)
```
