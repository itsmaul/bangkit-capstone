```python
import json
import numpy as np
import pandas as pd
```

```python
def preprocess_data(data):
    # Menambahkan kolom untuk kategori income dan expense berdasarkan transaction_type
    data['income'] = data.apply(lambda x: x['amount'] if x['transaction_type'] == 'Income' else 0, axis=1)
    data['expense'] = data.apply(lambda x: x['amount'] if x['transaction_type'] == 'Expense' else 0, axis=1)

    # Memisahkan data untuk income dan expense
    income_df = data[data['transaction_type'] == 'Income'][['date', 'amount']].copy()
    expense_df = data[data['transaction_type'] == 'Expense'][['date', 'amount']].copy()

    # Menghapus baris yang tidak memiliki nilai (nilai 0) pada masing-masing DataFrame
    income_df = income_df[income_df['amount'] > 0]
    expense_df = expense_df[expense_df['amount'] > 0]

    return income_df, expense_df
```

```python
merger_df = pd.read_json("merger_dataset.json")
```

```python
income_df, expense_df = preprocess_data(merger_df)
```

```python
# Simpan 'income_df' ke file JSON
with open('income_data.json', 'w') as json_file:
    income_df['date'] = pd.to_datetime(income_df['date'])
    income_df['date'] = income_df['date'].dt.strftime('%Y-%m-%d')
    json.dump(income_df.to_dict(orient="records"), json_file, indent=1)

# Simpan 'expense_df' ke file JSON
with open('expense_data.json', 'w') as json_file:
    expense_df['date'] = pd.to_datetime(expense_df['date'])
    expense_df['date'] = expense_df['date'].dt.strftime('%Y-%m-%d')
    json.dump(expense_df.to_dict(orient="records"), json_file, indent=1)
```
